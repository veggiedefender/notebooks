```python
from math import gcd
import matplotlib.pyplot as plt
```

```python
def U(n):
    return [a for a in range(n) if gcd(a, n) == 1]
```

```python
def cayley_table(n):
    u = U(n)
    return [[(a * b) % n for b in u] for a in u]
```

```python
cayley_table(20)
```

```python
plt.imshow(cayley_table(20))
```

```python
plt.imshow(cayley_table(37))
```

```python
plt.imshow(cayley_table(103))
```
